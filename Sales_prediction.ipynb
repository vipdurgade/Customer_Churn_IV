```python
import pandas as pd

# Load the CSV with German encoding
df = pd.read_csv("D:/IV_Mastertheis/Merged_kd_data.csv", encoding='latin1')

# List of columns to convert
columns_to_convert = ['gesbeitr', 'SCHAGESA', 'KOSTGESA', 'KOSTGEKA', 'kosten_verw', 'kosten_prov', 'SDBEITR5', 'KILOMETERSTAND']

# Function to convert German number format to float
def convert_german_number(val):
    # Handle NaN, None, or non-string types
    if pd.isna(val) or val is None:
        return None
    # Convert to string to handle numeric or mixed types
    val = str(val).strip()
    # Replace German number format (e.g., "1.234,56" -> "1234.56")
    val = val.replace('.', '').replace(',', '.')
    try:
        return float(val)
    except ValueError:
        return None  # Return None for non-numeric values

# Apply the function to the selected columns
for col in columns_to_convert:
    if col in df.columns:
        df[col] = df[col].apply(convert_german_number)

# Optional: Save to a new file
# df.to_csv("converted_kd_2022.csv", index=False)

# Display first few rows to confirm
print(df[columns_to_convert].head())
```

```python


# Convert both columns to datetime (German format)
df['beginn'] = pd.to_datetime(df['beginn'], format='%d.%m.%Y', errors='coerce')
df['datum_s'] = pd.to_datetime(df['datum_s'], format='%d.%m.%Y', errors='coerce')

# Fill NaN values in datum_s with specific date
fill_date = pd.to_datetime("2025-05-26")
df['datum_s'] = df['datum_s'].fillna(fill_date)

# Now calculate the difference in days
df['duration_days'] = (df['datum_s'] - df['beginn']).dt.days
```
